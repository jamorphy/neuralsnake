```python
import numpy as np
import matplotlib.pyplot as plt
import json
from pathlib import Path
```

```python
episode_path = Path("collected_data/episode_000000")
frames = np.load(episode_path / "frames.npy")
with open(episode_path / "episode_data.json", "r") as f:
    metadata = json.load(f)
```

```python
print("Episode stats:")
print(f"Number of frames: {len(frames)}")
print(f"Frame shape: {frames.shape}")
print(f"Total reward: {metadata['total_reward']}")
print(f"Episode length: {metadata['episode_length']}")
```

```python
# Display first few frames
fig, axes = plt.subplots(1, 4, figsize=(20, 5))
for i in range(4):
    axes[i].imshow(frames[i])
    axes[i].axis('off')
    axes[i].set_title(f'Frame {i}')
plt.tight_layout()
plt.show()
```

```python
# Display a sequence of frames with actions
start_idx = 200  # Start from frame 10
num_frames = 5  # Show 5 consecutive frames

fig, axes = plt.subplots(1, num_frames, figsize=(25, 5))
for i in range(num_frames):
    idx = start_idx + i
    axes[i].imshow(frames[idx])
    axes[i].axis('off')
    action = metadata['actions'][idx]
    axes[i].set_title(f'Frame {idx}\nAction: {action}')
plt.tight_layout()
plt.show()
```

```python
# Print action distribution
from collections import Counter
action_counts = Counter(metadata['actions'])
print("\nAction distribution:")
for action, count in action_counts.items():
    print(f"Action {action}: {count} times")
```
